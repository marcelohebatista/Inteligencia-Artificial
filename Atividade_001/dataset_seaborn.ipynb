# 📚 Datasets do Seaborn — Descrição (catálogo completo)

A lista abaixo contém **todos os 22 datasets** disponíveis via `sns.get_dataset_names()` no Seaborn, com uma explicação curta e uma sugestão de uso em aula (EDA/ML).

---

## 1) `anagrams`
- **O que é:** dados de experimentos relacionados a **anagramas** (tarefa cognitiva/linguagem).
- **Bom para:** EDA, comparação entre condições/grupos.
- **ML (se quiser):** classificação/regressão simples dependendo da variável alvo escolhida (normalmente é mais didático para estatística/visual).

## 2) `anscombe`
- **O que é:** o **Quarteto de Anscombe** (vários datasets com estatísticas quase iguais, mas gráficos muito diferentes).
- **Bom para:** lição clássica: *“sempre plote os dados”*.
- **ML:** não é foco; é uma aula de “não confie só em média e variância”.

## 3) `attention`
- **O que é:** dados de **experimentos de atenção** (respostas sob condições diferentes).
- **Bom para:** EDA, boxplots, comparação de grupos, testes simples.
- **ML:** pode virar classificação/regressão dependendo do target, mas o valor maior é a interpretação.

## 4) `brain_networks`
- **O que é:** dados de **redes cerebrais** (normalmente matrizes/medidas entre regiões).
- **Bom para:** visualização de matrizes, heatmaps, correlação e redes.
- **ML:** avançado; pode virar classificação por sujeito/condição se você preparar features.

## 5) `car_crashes`
- **O que é:** estatísticas de **acidentes de carro** (por estado/região, etc.).
- **Bom para:** regressão e EDA com variáveis contínuas.
- **ML:** regressão (prever uma taxa/indicador a partir de outros).

## 6) `diamonds`
- **O que é:** características de diamantes (**carat, cut, color, clarity...**) e **preço**.
- **Bom para:** pipeline com **categóricas + numéricas**, one-hot e regressão.
- **ML:** **regressão** (target clássico: `price`).

## 7) `dots`
- **O que é:** dataset **sintético** para demonstração de gráficos.
- **Bom para:** visualização (lineplot/scatter), noções de tendências.
- **ML:** não é um benchmark real; é mais “dataset de demonstração”.

## 8) `dowjones`
- **O que é:** série/medidas relacionadas ao **mercado financeiro** (Dow Jones).
- **Bom para:** EDA temporal, tendência, variação.
- **ML:** regressão/séries temporais (nível 2: exige cuidados com *time split*).

## 9) `exercise`
- **O que é:** dados de **exercício físico** (ex.: pulso/tempo/dieta/atividade, dependendo das colunas).
- **Bom para:** EDA + comparação entre categorias.
- **ML:** regressão ou classificação dependendo do target que você definir.

## 10) `flights`
- **O que é:** número de passageiros por **mês/ano** (série temporal clássica).
- **Bom para:** pivot, heatmap, sazonalidade.
- **ML:** séries temporais (nível 2). Em ML tabular, exige transformação.

## 11) `fmri`
- **O que é:** medições de sinais (fMRI) ao longo do tempo e condições.
- **Bom para:** gráficos com intervalos/confiança, séries por condição.
- **ML:** possível, mas normalmente é dataset para visualização/estatística experimental.

## 12) `geyser`
- **O que é:** dados do gêiser **Old Faithful** (duração/intervalo).
- **Bom para:** distribuição, correlação, **clusterização** simples.
- **ML:** regressão (ex.: prever intervalo) ou clustering (KMeans).

## 13) `glue`
- **O que é:** medições experimentais (ex.: força/tempo/variáveis de “cola”).
- **Bom para:** EDA e regressão simples.
- **ML:** regressão (pequeno, ótimo para treino rápido).

## 14) `healthexp`
- **O que é:** dados de **saúde/expectativa de vida** ao longo do tempo (geralmente por país/ano).
- **Bom para:** séries temporais, comparação por grupos.
- **ML:** regressão temporal (nível 2), ou regressão tabular após feature engineering.

## 15) `iris`
- **O que é:** dataset clássico de flores (medidas das pétalas/sépalas + espécie).
- **Bom para:** introdução de classificação, visualização 2D/3D, baseline rápido.
- **ML:** **classificação multiclasse** (target: `species`).

## 16) `mpg`
- **O que é:** dados de carros (peso, cilindros etc.) e consumo **mpg**.
- **Bom para:** regressão + mistura de numéricas/categóricas.
- **ML:** **regressão** (target: `mpg`).

## 17) `penguins`
- **O que é:** medições de pinguins + variáveis categóricas (ilha/sexo/espécie).
- **Bom para:** “Iris moderno” com mais realidade (categóricas e possivelmente missing).
- **ML:** **classificação multiclasse** (target: `species`).

## 18) `planets`
- **O que é:** dados de **exoplanetas** (método de detecção, massa, distância etc.) com muitos faltantes.
- **Bom para:** treinar **imputação** + pipeline, análise de missing.
- **ML:** classificação (target: `method`) ou regressão (ex.: `mass`, se fizer sentido após limpeza).

## 19) `seaice`
- **O que é:** medidas de **gelo marinho** ao longo do tempo.
- **Bom para:** visualização temporal, tendências e sazonalidade.
- **ML:** séries temporais (nível 2).

## 20) `taxis`
- **O que é:** corridas de táxi (tipicamente com tempo/local/contagens).
- **Bom para:** EDA com tempo + categorias; pode ser maior.
- **ML:** regressão (prever fare/duração) ou classificação (ex.: período/demanda), dependendo de como você define o target.

## 21) `tips`
- **O que é:** dados de restaurante: conta total, gorjeta, fumante, dia, horário, sexo etc.
- **Bom para:** pipeline (categorias) e interpretação de métricas.
- **ML:** regressão (target: `tip`) ou classificação (target: `smoker`).

## 22) `titanic`
- **O que é:** passageiros do Titanic (features numéricas e categóricas + missing).
- **Bom para:** pipeline completo (imputer + one-hot) + avaliação com matriz de confusão.
- **ML:** **classificação binária** (target: `survived`).
